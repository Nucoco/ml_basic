```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# k-means( clustering)
df = pd.read_csv('convinience_store.csv')
df
```

```python
x = df.drop('No', axis=1).values
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3, random_state=0)
kmeans.fit(x)
```

```python
# center position in each column
kmeans.cluster_centers_
# the above will show 6*3( columns * clusters)
```

```python
cluster = kmeans.predict(x)
cluster
```

```python
df_cluster = df.copy()
df_cluster['cluster'] = cluster
df_cluster.head()
```

```python
# In Unsupervised learning, the answer does not exist unlike supervised learning.
# So, humans need to consider how to utilize the outputs and what to utilize the outputs for. 
```

```python
# consideration with the mean of total amount in each cluster
# To process clustered data
df_results = pd.DataFrame()
df_cluster[df_cluster['cluster'] == 0].mean()
```

```python
df_results['cluster 0'] = df_cluster[df_cluster['cluster'] == 0].mean().tolist()
df_results['cluster 1'] = df_cluster[df_cluster['cluster'] == 1].mean().tolist()
df_results['cluster 2'] = df_cluster[df_cluster['cluster'] == 2].mean().tolist()
```

```python
df_results
```

```python
df_results = df_results.set_index(df_cluster.columns)
df_results
```

```python
df_results = df_results.drop(['No', 'cluster']).T
df_results
```
