```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_boston
```

```python
dataset = load_boston()
dataset
```

```python
x, t = dataset.data, dataset.target
columns = dataset.feature_names
```

```python
type(x), x.shape, type(t), t.shape, columns
```

```python
df = pd.DataFrame(x, columns=columns)
df.head()
```

```python
df['Target'] = t
df.head()
```

```python
"""
Data processing: use DataFrame (pandas)
Use of sklearn: transform to ndarray (Numpy) => df[''].values
"""
```

```python
t = df['Target'].values
x = df.drop(labels=['Target'], axis=1).values
```

```python
# split data into for train and for test (hold out method)
from sklearn.model_selection import train_test_split
```

```python
train_test_split(x, t, test_size=0.3, random_state=0)
```

```python
x_train, x_test, t_train, t_test = train_test_split(x, t, test_size=0.3, random_state=0)
```

```python
len(x_train), len(x_test)
```

```python
# 1 define a model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
# 2 train a model
model.fit(x_train, t_train)
```

```python
# In multiple regression, parameters are the weight and the bias

# confirm a trained data model( weights for each input variables 13)
model.coef_
```

```python
# visualization
plt.figure(figsize=(10, 7))
plt.bar(x=columns, height=model.coef_)
```

```python
# bias
model.intercept_
```

```python
# 3 validation ( by using score(): returns Coefficient of determination)
print(f'train score: {model.score(x_train, t_train)}')
print(f'test score: {model.score(x_test, t_test)}')
```

```python
"""overfitting: a descrepency is between train score and test score"""
```

```python
# 4 inference( input a new value to a trained model and let it predict a value)
y = model.predict(x_test)

print(f'Predicted value: {y[0]}')
print(f'Target value: {t_test[0]}')
```

```python
"""
Correlation & Multicollinearity
"""
df = pd.read_csv('regression_pls.csv')
df.shape
```

```python
# REVIEW
# transform to ndarray
t = df['Target'].values
x = df.drop('Target', axis=1).values
t.shape, x.shape
```

```python
# divide data into train and test
x_train, x_test, t_train, t_test = train_test_split(x, t, test_size=0.3, random_state=0)
```

```python
# algorithm
model = LinearRegression()
# train
model.fit(x_train, t_train)
# parameter
# print(f'Weight: {model.coef_}, Bias: {model.intercept_}')
# validation
print(f'train score: {model.score(x_train, t_train)}')
print(f'test score: {model.score(x_test, t_test)}')
```

```python
# overfitting
# - in this case, Multicollinearity is happen.
# - It's triggered by input variables that are high Correlation.
```

```python
# Check Correlation between input variables
df_corr = df.corr()
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(df_corr.iloc[:20, :20], annot=True)
```

```python
sns.jointplot(x='x1', y='x16', data=df)
```

```python
# PLS(Partial Least Squares)
#  - this is like 'Principal component analysis', 'Dimensionality reduction'
#  - n_components: Latent variable( hyper parameter)
from sklearn.cross_decomposition import PLSRegression

# explore a better hyper parameter's value to create a better trained model
pls = PLSRegression(n_components=11)
pls.fit(x_train, t_train)

print(f'train score: {pls.score(x_train, t_train)}')
print(f'test score: {pls.score(x_test, t_test)}')
```
